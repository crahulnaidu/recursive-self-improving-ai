```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
X=2*np.random.randn(100,1)
Y=3+4*X+np.random.randn(100,1)

X_b=np.c_[np.ones((100,1)),X]

def compute_cost_function(theta):
  m=len(Y)
  return (1/(2*m))*np.sum((X_b.dot(theta)-Y)**2)


theta_arr=np.linspace(-10,10,100)
cost_vals=[]

for t in theta_arr:
  theta=np.array([[0],[t]])
  cost_vals.append(compute_cost_function(theta))


plt.plot(theta_arr,cost_vals)
plt.xlabel("theta values")
plt.ylabel("cost value for theta")
plt.title("Cost function landscape")
plt.show()
```

```python
print(min(cost_vals))
```

```python
theta = np.random.randn(2,1)
lr = 0.01
epochs = 50

theta_history = []

for i in range(epochs):
    gradients = 2/100 * X_b.T.dot(X_b.dot(theta) - Y)
    theta = theta - lr * gradients
    theta_history.append(theta[1][0])

# Plot path
plt.plot(theta_arr, cost_vals)
plt.scatter(theta_history, [compute_cost_function(np.array([[0],[t]])) for t in theta_history], color='red')
plt.title("Gradient Descent Path")
plt.show()
```

```python



def compute_cost(theta):
    m = len(Y)
    return (1/(2*m)) * np.sum((X_b.dot(theta) - Y)**2)



def gradient_descent(theta, lr, epochs):
    theta_history = []

    for _ in range(epochs):
        gradients = 2/len(Y) * X_b.T.dot(X_b.dot(theta) - Y)
        theta = theta - lr * gradients

        theta_history.append(theta.copy())

    return theta, theta_history


theta_vals = np.linspace(-10, 10, 200)
cost_vals = []

for t in theta_vals:
    theta_temp = np.array([[0], [t]])
    cost_vals.append(compute_cost(theta_temp))



learning_rates = [0.01, 0.1, 0.5]
colors = ['blue', 'green', 'red']

plt.figure(figsize=(10,6))



plt.plot(theta_vals, cost_vals, label="Cost Function")



for lr, color in zip(learning_rates, colors):
    theta_init = np.random.randn(2,1)

    _, theta_history = gradient_descent(theta_init, lr, 50)

    theta_path = [t[1][0] for t in theta_history]
    cost_path = [compute_cost(np.array([[0],[t]])) for t in theta_path]

    plt.plot(theta_path, cost_path, marker='o', color=color, label=f"lr={lr}")


plt.xlabel("Theta (slope)")
plt.ylabel("Cost")
plt.title("Gradient Descent Explorer")
plt.legend()
plt.show()
```
