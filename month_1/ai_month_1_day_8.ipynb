```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
np.random.seed(0)
X=np.random.randn(100,2)
Y=(X[:,0]+X[:,1]>0).astype(int).reshape(-1,1)

X_b=np.c_[np.ones((100,1)),X]
theta=np.random.rand(3,1)
lr=0.1

for _ in range(1000):
  z=X_b @ theta
  y_pred=sigmoid(z)

  gradient=(1/100)*X_b.T @ (y_pred-Y)
  theta=theta-lr*gradient


print("theta values",theta)
```

```python
def predict(X):
  y_pred=sigmoid(X @ theta)
  return (y_pred>=0.5).astype(int)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y.flatten())
plt.title('Data')
plt.show()
```

```python
#comparing linear and logistic regression for classification.
```

```python
model=LinearRegression()
model.fit(X,Y)

X_test=np.random.randn(200,2)
y_test=(X_test[:,0]+X_test[:,1]>0).astype(int).reshape(-1,1)
```

```python
y_linear_predict=(model.predict(X_test)>=0.5).astype(int)

X_test_b=np.c_[np.ones((200,1)),X_test]
y_logistic=predict(X_test_b)
```

```python
linear_error=np.mean(np.abs(y_linear_predict-y_test))
logistic_error=np.mean(np.abs(y_logistic-y_test))
```

```python
print("error due to linear regression",linear_error)
print("error due to logistic regression",logistic_error)
```
