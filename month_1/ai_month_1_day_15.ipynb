```python
import numpy as np
import matplotlib.pyplot as plt

#simple quadratic loss

theta0=np.linspace(-5,5,100)
theta1=np.linspace(-5,5,100)

T0,T1=np.meshgrid(theta0,theta1)

#loss function

J=T0**2+5*T1**2

plt.contour(T0,T1,J,levels=30)
plt.xlabel("theta0")
plt.ylabel("theta1")
plt.title("Loss landscape(Bad Conditioning)")
plt.show()
```

```python
theta=np.array([4.0,4.0])
lr=0.1


path=[theta.copy()]

for _ in range(20):
  grad=np.array([2*theta[0],10*theta[1]])
  theta=theta-lr*grad
  path.append(theta.copy())

path=np.array(path)

plt.contour(T0,T1,J,levels=30)
plt.plot(path[:,0],path[:,1],'ro--')
plt.title("Gradient Descent path")
plt.show()
```
