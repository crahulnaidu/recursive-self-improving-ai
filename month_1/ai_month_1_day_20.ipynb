```python
import numpy as np

#Inputs
x=2.0
w=0.5
b=0.1
y_true=1.0

#Forward
z=x*w+b
a=1/(1+np.exp(-z))
loss=0.5*(y_true-a)**2

#Backward propogation

dL_da=a-y_true
da_dz=a*(1-a)
dz_dw=x

dL_dw=dL_da*da_dz*dz_dw

print("Gradient wrt w:",dL_dw)
```
